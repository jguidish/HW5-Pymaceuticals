```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

#Compare
drug_list= ['Capomulin','Infubinol','Ketapril','Placebo','Stelasyn','Naftisol','Ceftamin','Propriva','Zoniferol','Ramicane']
markers= ['o','x','d','^']

#Read
mouse_drug_data=pd.read_csv(mouse_drug_data_to_load, low_memory=False)
clinical_trial_data=pd.read_csv(clinical_trial_data_to_load, low_memory=False)

#Combine
clinical_drug_trial_df=pd.merge(clinical_trial_data,mouse_drug_data, how='left', on='Mouse ID')

#Display
clinical_drug_trial_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
drug_timepoint=clinical_drug_trial_df.groupby(['Drug','Timepoint'])
mean_tumor_vol=drug_timepoint['Tumor Volume (mm3)'].mean()

# Convert to DataFrame
mean_tumor_vol_df=pd.DataFrame(mean_tumor_vol)

#Print
mean_tumor_vol_df=mean_tumor_vol_df.reset_index()
print(mean_tumor_vol_df.keys())

# Preview DataFrame
mean_tumor_vol_df.head()
```

```python
stem_tumor_vol=drug_timepoint['Tumor Volume (mm3)'].sem()

#Convert
stem_tumor_vol_df=pd.DataFrame(stem_tumor_vol)
stem_tumor_vol_df=stem_tumor_vol_df.reset_index()

#Print
print(stem_tumor_vol_df.keys())
stem_tumor_vol_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_tumor_vol_df_pivot=mean_tumor_vol_df.pivot(index='Timepoint',columns='Drug',values='Tumor Volume (mm3)')
stem_tumor_vol_df_pivot=stem_tumor_vol_df.pivot(index='Timepoint',columns='Drug',values='Tumor Volume (mm3)')

# Preview that Reformatting worked
mean_tumor_vol_df_pivot
```

```python
#Standard Error of Tumor Vol
stem_tumor_vol_df_pivot=stem_tumor_vol_df.pivot(index='Timepoint',columns='Drug',values='Tumor Volume (mm3)')
stem_tumor_vol_df_pivot
```

```python
# Generate the Plot (with Error Bars)
fig, ax=plt.subplots()
mean_vol=mean_tumor_vol_df_pivot
stem_vol=stem_tumor_vol_df_pivot

mean_timepoint=mean_vol.index

i=0
for drug in drug_list:
        ax.errorbar(mean_timepoint, mean_vol[drug], yerr=stem_vol[drug],marker=markers[i],color=colors[i])
        i+=1
        
plt.legend=("best")

i=0
for drug in drug_list:
        plt.plot(mean_timepoint, mean_vol[drug], color=colors[i])
        i+=1
        
plt.title('Tumor Response to Treatment')
plt.xlabel("Timepoint (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.grid()
        
        
# Save the Figure
plt.savefig('Tumor_Response.png', dpi=300, format='png',bbox_inches='tight')
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
mean_met_sites=drug_timepoint['Metastatic Sites'].mean()

# Convert to DataFrame
mean_met_sites_df=pd.DataFrame(mean_met_sites)
mean_met_sites_df=mean_met_sites_df.reset_index()

# Preview DataFrame
mean_met_sites_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
stderr_sites=drug_timepoint['Metastatic Sites'].sem()

# Convert to DataFrame
stderr_sites_df=pd.DataFrame(stderr_sites)
stderr_sites_df=stderr_sites_df.reset_index()

# Preview DataFrame
stderr_sites_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_met_sites_df_pivot=mean_met_sites_df.pivot(index='Timepoint',columns='Drug',values='Metastatic Sites')

# Preview that Reformatting worked
mean_met_sites_df_pivot
```

```python
stderr_sites_df_pivot=stderr_sites_df.pivot(index='Timepoint',columns='Drug',values='Metastatic Sites')
stderr_sites_df_pivot
```

```python
# Generate the Plot (with Error Bars)
fig, ax=plt.subplots()
mean_sites=mean_met_sites_df_pivot
stem_sites=stderr_sites_df_pivot

mean_timepoint=mean_sites.index

i=0
for drug in drug_list:
    ax.errorbar(mean_timepoint, mean_sites[drug], yerr=stem_sites[drug],marker=markers[i],color=colors[i])
    i+=1
    
plt.legend(loc="best")

i=0
for drug in drug_list:
    plt.plot(mean_timepoint, mean_sites[drug],color=colors[i])
    i+=1
    
plt.grid()
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration Days")
plt.ylabel("Metastatic Sites")

# Save the Figure
plt.savefig('Meta_Spread.png',dpi=300, format='png', bbox_inches='tight')

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mice_count=drug_timepoint['Mouse ID'].count()
mice_count

# Convert to DataFrame
mice_count_df=pd.DataFrame(mice_count)
mice_count_df=mice_count_df.reset_index()

# Preview DataFrame
mice_count_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mice_count_df_pivot=mice_count_df.pivot(index='Timepoint',columns='Drug',values='Mouse ID')

# Preview the Data Frame
mice_count_df_pivot
```

```python
# Generate the Plot (Accounting for percentages)
mice_count_timepoint=mice_count_df_pivot.index
i=0
for drug in drug_list:
    plt.plot(mice_count_timepoint, (mice_count_df_pivot[drug]/mice_count_df_pivot[drug].max()*100), marker=markers[i], linestyle='--',color=colors[i])
    i+=1

plt.grid()
plt.legend(loc="best")
plt.title("Survival During Treatment")
plt.xlabel("Timepoint (Days)")
plt.ylabel("Survival Rate %")

# Save the Figure
plt.savefig('Survival_Rate.png', dpi=300, format='png', bbox_inches='tight')

# Show the Figure
plt.show()
```
